Foreign Soccer Players in CONCACAF Leagues and CONCACAF Talent

```python
import pandas as pd
import seaborn as sns
import numpy as np

%matplotlib inline
```

```python
fifa = pd.read_csv('C:/Users/rp/Documents/FIFA Full Data2.csv')
fifa.head()
```

```python
fifa['league'].unique()
```

```python
def leagues(lea):
    if lea == 'MLS':
        return 'USA'
    else:
        return lea

fifa['league']=fifa['league'].apply(leagues)
```

```python
fifa['league'].value_counts()
```

```python
players = fifa.dropna()
```

```python
players.head()
```

```python
concacaf = fifa[(fifa['nationality'] == 'Honduras') | (fifa['nationality'] == 'Costa Rica') | 
                (fifa['nationality'] == 'Mexico') | (fifa['nationality'] == 'Panama') |
                (fifa['nationality'] == 'United States')]
```

```python
import matplotlib.pyplot as plt
```

```python
g = sns.stripplot(x='league', y='rating', data=concacaf, hue='nationality', palette='plasma')
g.set_xlabel('League')
g.set_ylabel('Rating')
g.set_title('CONCACAF Soccer Players and Leagues')
plt.legend(loc='center left', bbox_to_anchor=(1.0,0.5), title='Nationality')
```

```python
s = sns.barplot(x='nationality',y='rating',data=concacaf,estimator=np.mean, palette='plasma')
s.set_xlabel('Nationality')
s.set_ylabel('Average Rating')
s.set_title('Top CONCACAF Teams and Average FIFA Player Rating')
s.savefig('CONCACAF RATING.png')
s.s
```
