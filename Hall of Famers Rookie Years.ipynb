```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
batters = pd.read_csv('Hall of Famers Rookie Years.csv')
```

```python
batters.info()
```

```python
batters['Debut'] = pd.to_datetime(batters['Debut'], format = '%Y-%m-%d')
#batters['Debut'].astype('datetime')
```

```python
batters['Debut'] = batters['Debut'].dt.year
batters.head()
```

```python
batters2 = batters[(batters['Debut'] == batters['Year'])]
```

```python
batters2.head()
```

```python
len(batters2)
```

```python
betterrookies = batters2[batters2['Rookie Batting Average'] > batters2['Career Batting Average']]
betterrookies
```
